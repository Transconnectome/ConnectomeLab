## GPS with PCA
- I will combine two data: GPS_total and PCA
- Remove all the NA
- train / test set will be seperated (4:1)

```python
import pandas as pd
```

```python
pca = pd.read_csv('ABCD_QCed_EA_morePruned_pca.eigenvec', sep='\s', header = None)
pca.head()
```

```python
pca = pca.drop(1, axis = 1)
pca.head()
```

```python
pca.columns = ['KEY', 'PC1','PC2',  'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10']
pca.head()
```

```python
pca['KEY'] = ["".join(key.split('_')[1:3]) for key in pca['KEY']]
pca.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import r2_score
```

```python
gps_phe = pd.read_csv('/home/ubuntu/gps_and_phe_nihtotcomp_notnull.csv', header=0, index_col=0)
gps_phe.head()
```

```python
gps_phe.columns
```

```python
gps_pca_phe = gps_phe.join(pca.set_index('KEY'), how='inner')
gps_pca_phe
```

```python
X = gps_pca_phe.iloc[:,67:77]
X = pd.DataFrame(gps_pca_phe.iloc[:,0:25], X)
features = X.columns
X.head()
```

```python
y = gps_pca_phe['nihtbx_totalcomp_uncorrected']
```

```python
X.to_numpy()
y.to_numpy().ravel()
```

```python
k_fold = 5
max_depth = [3, 4, 5, 6]
n_estimators = [100]

params={'max_depth': max_depth, 'n_estimators': n_estimators}
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train[X_train['nihtbx_totalcomp_uncorrected'].notnull()]
X_train.shape
```

```python
rf = RandomForestRegressor()
grid_cv = GridSearchCV(estimator = rf, param_grid = params, cv = k_fold, scoring='r2', n_jobs=-1, return_train_score=True)
```

```python
grid_cv.fit(X_train, y_train)
y_pred = grid_cv.prediction(X_test)
```
